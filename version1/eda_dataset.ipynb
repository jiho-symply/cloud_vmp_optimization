```python
import os
import numpy as np
import pandas as pd

pd.set_option('display.float_format', '{:.6f}'.format)

df = pd.read_csv("sample_vm_data.csv")
df
```

```python
vmtable = pd.read_csv("trace_data/vmtable/vmtable.csv", header=None)
vmtable.columns = ["vm_id", "subscription_id", "deployment_id",
                   "timestamp_vm_created", "timestamp_vm_deleted",
                   "max_cpu", "avg_cpu", "p95_max_cpu", "vm_category",
                   "vm_virtual_core_count_bucket", "vm_memory_gb_bucket"]

vmtable["duration"] = vmtable["timestamp_vm_deleted"] - vmtable["timestamp_vm_created"]
vmtable["duration"] /= 3600

vmtable
```

```python
vmtable = vmtable.loc[vmtable["duration"] >= 1]
vmtable.reset_index(drop=True)
```

```python
print(len(vmtable))
vmtable.loc[:, :"deployment_id"].nunique()
```

```python
vmtable.groupby("deployment_id", as_index=False).agg('size').sort_values('size')
```

```python
(
    vmtable
    .groupby(["deployment_id","subscription_id","vm_category"], as_index=False)
    .agg(
        size=('max_cpu','size'), 
        avg_dur=('duration', 'mean'), 
        max_dur=('duration', 'max'),
        avg_cpu=('avg_cpu', 'mean'),
        p95_max_cpu=('p95_max_cpu', 'mean'),
    )
    .sort_values('size')
).tail(20)
```

```python
(
    vmtable
    .groupby(["vm_virtual_core_count_bucket"], as_index=False)
    .agg(
        size=('max_cpu','size'), 
        avg_dur=('duration', 'mean'), 
        max_dur=('duration', 'max'),
        avg_cpu=('avg_cpu', 'mean'),
        p95_max_cpu=('p95_max_cpu', 'mean'),
    )
    .sort_values('size')
).tail(20)
```

```python
(
    vmtable
    .groupby(["vm_virtual_core_count_bucket"]).value_counts(["vm_category"])
    .sort_index()
    .reset_index()
).tail(20)
```

```python
(
    vmtable
    .groupby(["deployment_id","vm_virtual_core_count_bucket"], as_index=False)
    # .groupby(["deployment_id","subscription_id"], as_index=False)
    # .groupby(["subscription_id"], as_index=False)
    .agg(
        size=('vm_category','size'),
        delay_insensitive=('vm_category',lambda x: sum(x=="Delay-insensitive")),
        interactive=('vm_category',lambda x: sum(x=="Interactive")),
        unknown=('vm_category',lambda x: sum(x=="Unknown")),
    )
    .sort_values('size')
).tail(30)
```
